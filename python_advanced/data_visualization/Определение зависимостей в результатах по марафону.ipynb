```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

def str_time_to_secs(s):
    pt = datetime.strptime(s,'%H:%M:%S')
    total_seconds = pt.second + pt.minute*60 + pt.hour*3600

    return total_seconds

data = pd.read_csv('https://video.ittensive.com/python-advanced/marathon-data.csv')
data['split'] = data['split'].apply(lambda x: str_time_to_secs(x))
data['final'] = data['final'].apply(lambda x: str_time_to_secs(x))

# print(data['age'].unique())

# sns.pairplot(data, hue='gender', height=4)
# plt.show()
```

```python
print('Correlation for')
print('split VS final')
print(round(stats.pearsonr(data['split'], data['final'])[0], 2))
# print('age VS split')
# print(round(stats.pearsonr(data['age'], data['split'])[0], 2))
# print('age VS final')
# print(round(stats.pearsonr(data['age'], data['final'])[0], 2))

sns.jointplot(x='split', y='final', data=data, height=6, kind='kde')
plt.show()
```
